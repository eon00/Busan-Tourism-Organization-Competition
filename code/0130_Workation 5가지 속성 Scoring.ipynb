```python
# 접근성 9% (버스정류장 0.045, 지하철역 0.045)
# 숙박유형 43%(외국인관광도시민박업 0.43*0.12, 일반야영장업 0.43*0.04, 한옥체험업 0.43*0.07, 관광펜션업 0.43*0.1, 관광숙박업 0.43*0.67)
# IT 7%(와이파이 0.07)
# 식음 및 부대시설 20% (0.2)
# 퇴근 후 컨텐츠 21% (농어촌체험휴양마을 0.21*0.23, 관광명소 0.21*0.23, 레저사업장 수 0.21*0.09, 스포츠_체육_레저 수 0.21*0.09, 취미여가상품 0.21*0.36)

df8['성장가능성_SCORE'] = sum([df8.loc[:, '버스정류장 수'] * 0.045, df8.loc[:, '지하철역 수'] * 0.045, # 접근성
                            df8.loc[:, '외국인관광도시민박업 수'] * 0.43 * 0.12, df8.loc[:, '일반야영장업 수'] * 0.43 * 0.04, df8.loc[:, '한옥체험업 수'] * 0.43 * 0.07,
                            df8.loc[:, '관광펜션업 수'] * 0.43 * 0.1, df8.loc[:, '관광숙박업 수'] * 0.43 * 0.67, # 숙박유형
                            df8.loc[:, '와이파이 수'] * 0.07 * 0.9, # IT
                            df8.loc[:, '식음료_편의시설 수'] * 0.2, # 식음 및 부대시설
                            df8.loc[:, '농어촌체험휴양마을 수'] * 0.21 * 0.23, df8.loc[:, '관광명소 수'] * 0.21 * 0.23, df8.loc[:, '레저사업장 수'] * 0.21 * 0.09,
                            df8.loc[:, '스포츠_체육_레저 수'] * 0.21 * 0.09, df8.loc[:, '취미여가상품 수'] * 0.21 * 0.36]) # 퇴근 후 컨텐츠
```
